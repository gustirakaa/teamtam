```python
import json
import pandas as pd
data = []
with open('movies.json', 'r') as json_data:
  for line in json_data:
    data.append(json.loads(line))
```

```python
def normalize_numbers(obj):
    """Recursively normalize numbers in the JSON."""
    if isinstance(obj, dict):
        # Rekursi untuk setiap key dalam dictionary
        return {k: normalize_numbers(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        # Rekursi untuk setiap elemen dalam list
        return [normalize_numbers(i) for i in obj]
    elif isinstance(obj, (str, int, float)):
        # Jika tipe data string/int/float, return apa adanya
        return obj
    elif "$numberInt" in obj:
        # Jika ada $numberInt, konversi ke float
        return float(obj["$numberInt"])
    elif "$numberDouble" in obj:
        # Jika ada $numberDouble, konversi ke float
        return float(obj["$numberDouble"])
    elif "$date" in obj:
        # Jika ada $date, tetap return nilai waktu dalam long
        return obj["$date"]["$numberLong"] if isinstance(obj["$date"], dict) else obj["$date"]
    else:
        # Jika tipe data lain, return apa adanya
        return obj
```

```python
def normalize_numbers(obj):
    """Recursively normalize numbers in the JSON."""
    if isinstance(obj, dict):
        # Recursively process each key in the dictionary
        return {k: normalize_numbers(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        # Recursively process each element in the list
        return [normalize_numbers(i) for i in obj]
    elif isinstance(obj, (str, int, float)):
        # If the type is string/int/float, return it as is
        return obj
    elif "$numberInt" in obj:
        # If $numberInt is found, convert to float
        return float(obj["$numberInt"])
    elif "$numberDouble" in obj:
        # If $numberDouble is found, convert to float
        return float(obj["$numberDouble"])
    elif "$date" in obj:
        # If $date is found, return the long timestamp
        return obj["$date"]["$numberLong"] if isinstance(obj["$date"], dict) else obj["$date"]
    elif isinstance(obj, dict) and 'tomatoes' in obj:
        # Special case for tomatoes rating processing
        if 'viewer' in obj['tomatoes']:
            # Normalize viewer rating inside tomatoes
            obj['tomatoes']['viewer']['rating'] = float(obj['tomatoes']['viewer']['rating'])
        if 'critic' in obj['tomatoes']:
            # Normalize critic rating inside tomatoes
            obj['tomatoes']['critic']['rating'] = float(obj['tomatoes']['critic']['rating'])
        return obj
    else:
        # Return the object as is if no special handling is needed
        return obj
```

```python
normalized_data = [normalize_numbers(row) for row in data]
```

```python
df = pd.DataFrame(normalized_data)
```
